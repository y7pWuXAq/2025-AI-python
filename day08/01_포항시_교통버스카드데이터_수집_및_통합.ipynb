#### 영문 컬럼 이름을 한글명으로 변경하기

- 컬럼명 변경 순서
    - 1. 메타정의서에서 영문 및 한글 데이터가 있는 파일 읽어들이기
    - 2. 읽어들인 데이터를 딕셔너리 형태로 변환

```python
### 1. 메타 정의서에서 [영문] 및 [한글] 데이터가 있는 파일 읽어들이기
# - 읽어들일 컬럼명 : 컬럼명(영문), 컬럼명(한글) 2개만
# - 변수명 : df_bus_card_col_org

file_path2 = "./01_data_org/trfcard (0)/trfcard_columns.xlsx"

df_bus_card_col_org = pd.read_excel(file_path2,
                           header=2,
                           usecols="B:C")

df_bus_card_org.info() ## 컬럼 갯수 동일한지 확인
df_bus_card_col_org
```

```python
#### 2. 읽어들인 데이터를 딕셔너리 형태로 변환
# - key는 컬럼명(영문)
# - value는 컬럼명(한글)
# - 딕셔너리 변수명 : df_bus_card_col_dict
# - 결과 형태 : {'on_date' : "승차시각", 'off_date' : "하차시각", ... }

# 매칭 시킬 값 확인
# df_bus_card_col_org['컬럼명 (영문)'][1], df_bus_card_col_org['컬럼명 (한글)'][1]

# 빈 딕셔너리 생성
df_bus_card_col_dict = { }

# 각 행을 순차적으로 돌면서 영문 컬럼명과 한글 컬럼명으로 딕셔너리 생성
for i in range(len(df_bus_card_col_org)):
    # df_bus_card_col_org의 첫 번째 컬럼 (영문)과 두 번째 컬럼 (한글) 접근
    df_bus_card_col_dict[df_bus_card_col_org['컬럼명 (영문)'][i]] = df_bus_card_col_org['컬럼명 (한글)'][i]
    
df_bus_card_col_dict
```

```python
## 다른 방법

# zip(리스트1, 리스트2, ... 리스트n) : 모든 리스트의 갯수는 같아야 함
# - 각 리스트의 인덱스끼리 하나의 튜플로 묶는 작업
# - zip([a : a1], [b : b2]...)
# 결과 : [[a : b], [a1: b1], ... [an : bn]]

for k, v in zip(df_bus_card_col_org["컬럼명 (영문)"], df_bus_card_col_org["컬럼명 (한글)"]) :
    df_bus_card_col_dict[k] = v

df_bus_card_col_dict
```

```python
### 코드 줄이기

# zip() 함수 사용
df_bus_card_col_dict = {
    k : v for k, v in zip(df_bus_card_col_org["컬럼명 (영문)"],
                                 df_bus_card_col_org["컬럼명 (한글)"])}

df_bus_card_col_dict
```

```python
# 다른 방법

# to_numpy() : 데이터 프레임을 행렬의 리스트 형태로 변환하는 함수
# zip()의 구조와 유사 
# 왼쪽값을 key, 오른쪽 값을 value로 사용함

df_bus_card_col_dict = dict(df_bus_card_col_org.to_numpy())

df_bus_card_col_dict
```

```python
### 3. 원본 데이터 프레임의 컬럼명 변경하기 -> rename(columns=딕셔너리 변수) 함수 사용
# - 현재까지 원본 데이터 프레임 변수명 : df_bus_card_org
# - 결과 데이터 프레임 변수명 : df_bus_card_kor

df_bus_card_kor = df_bus_card_org.rename(columns=df_bus_card_col_dict)

df_bus_card_kor.head(1)
```

##### 소주제들에 필요한 컬럼 추가하기

```python
### 추가 컬럼 : 년, 월, 일, 시간, 분, 승객연령, 체류시간, 승하차구간(승차 + 하차 정류장)

## 냔, 월 ,일 ,시간, 분 -> 승차시각 데이터에서 추출
# - 승차시각 데이터 타입이 int ->  추출을 위해 str 타입으로 변환
# str 타입을 다시 날짜 타입으로 변환하여 필요한 값 추출

# - 변수 타입 변경(형변환) 함수 : int(), float(), str(), list(), dict(), tuple()
# - 데이터 프레임의 데이터 타입 변경 함수 : astype()
```

```python
### 승차시각 데이터를 문자 타입(object)으로 변환하기

df_bus_card_kor = df_bus_card_kor.astype({"승차시각" : str, "하차시각" : str})
df_bus_card_kor.info()
```

```python
pd.to_datetime(df_bus_card_kor["승차시각"])
```

```python
#### 승하차시각 데이터를 날짜 타입으로 변환하기
# - 날짜 타입으로 변환할 때는 형변환 형태가 아닌 (astype 함수 사용 안함) 날짜 라이브러리를 이용해서 값을 변환 후 넣어야 함
# - 날짜 라이브러리 : 데이터 프레임에서 제공하는 to_datetime() 함수 사용
# - 날짜 포멧 형태 : 전제 조건은 문자열인 경우
# yyyymmddhhmmss  //  yyyy-mm-dd hh:mm:ss // yyyy.mm.dd hh:mm:ss // yymmddhhmmss // yyyy/mm/dd hh:mm:ss

df_bus_card_kor["승차시각"] = pd.to_datetime(df_bus_card_kor["승차시각"])
df_bus_card_kor["하차시각"] = pd.to_datetime(df_bus_card_kor["하차시각"])

df_bus_card_kor.info()
```

```python
#### 원본 사용하지 않고, 새로운 데이터 프레임에 필요한 컬럼들만 별도 가공
# - 변수명 : df_bus_card
# - 별도 관리 컬럼 : 승차시각, 하차시각, 승객연령, 승차정류장, 하차정류장

df_bus_card = df_bus_card_kor[["승차시각","하차시각", "승객연령", "승차정류장", "하차정류장"]].copy()

df_bus_card.info()
```

```python
### 년, 월, 일, 시, 분 컬럼 추가하기

df_bus_card["기준년도"] = df_bus_card["승차시각"].dt.year
df_bus_card["기준월"] = df_bus_card["승차시각"].dt.month
df_bus_card["기준일"] = df_bus_card["승차시각"].dt.day
df_bus_card["기준시간"] = df_bus_card["승차시각"].dt.hour
df_bus_card["기준분"] = df_bus_card["승차시각"].dt.minute

df_bus_card.head(5)
```

```python
df_bus_card.info()
```

```python
#### "버스내체류시간(분)" 컬럼 추가
# - (하차시각 - 승차시각)을 분단위로 계산해서 처리
# -> 뺀 값을 초단위로 변환 -> 60으로 나눠주기

df_bus_card["버스내체류시간(분)"] = \
    round((df_bus_card["하차시각"] - df_bus_card["승차시각"]).dt.total_seconds() / 60, 2)

df_bus_card.head(2)
```

```python
#### 승하차구간 컬럼 추가하기
# - 승하차구간 = 승차정류장 + "->" + 하차정류장

df_bus_card["승하차구간"] = \
    df_bus_card_kor["승차정류장"] + " -> " + df_bus_card_kor["하차정류장"]

df_bus_card.head(2)
```

```python
import os
import pandas as pd

# 파일 경로
root = os.getcwd()
dir_nm = os.listdir(root)[0] # 01_data_org
file_list = os.listdir(dir_nm) # 01_data_org 안에 폴더 trfcard (0)

dir_nm
```

```python
import os
import pandas as pd

### 깨끗한 데이터프레임 정의하기
df_bus_card_tot = pd.DataFrame()

# 영문 컬럼명 -> 한글 컬럼명 정의
file_path2 = "./01_data_org/trfcard (0)/trfcard_columns.xlsx"

df_bus_card_col_org = pd.read_excel(file_path2,
                           header=2,
                           usecols="B:C")

#딕셔너리 형태로 저장 생성
df_bus_card_col_dict = dict(df_bus_card_col_org.to_numpy())


### file_list 만큼 반복하기
for file_nm in file_list :
    ### 파일 경로 정의하기
    file_path = f"./{dir_nm}/{file_nm}/trfcard.csv"
    # print(file_path) 

    try : 
        # 데이터 프레임의 컬럼명 변경하기
        df_bus_card_org = pd.read_csv(file_path)
        df_bus_card_kor = df_bus_card_org.rename(columns=df_bus_card_col_dict)
        # print(df_bus_card_kor) # rename 잘 됐는지 확인 
        
        # 승차시각 데이터를 문자 타입(object)으로 변환하기
        df_bus_card_kor = df_bus_card_kor.astype({"승차시각" : str, "하차시각" : str})

        # 승하차시각 데이터를 날짜 타입으로 변환하기
        df_bus_card_kor["승차시각"] = pd.to_datetime(df_bus_card_kor["승차시각"])
        df_bus_card_kor["하차시각"] = pd.to_datetime(df_bus_card_kor["하차시각"])

        # 원본 사용하지 않고, 새로운 데이터 프레임에 필요한 컬럼들만 별도 가공
        df_bus_card = df_bus_card_kor[["승차시각","하차시각", "승객연령", "승차정류장", "하차정류장"]].copy()

        # 년, 월, 일, 시, 분 컬럼 추가하기
        df_bus_card["기준년도"] = df_bus_card["승차시각"].dt.year
        df_bus_card["기준월"] = df_bus_card["승차시각"].dt.month
        df_bus_card["기준일"] = df_bus_card["승차시각"].dt.day
        df_bus_card["기준시간"] = df_bus_card["승차시각"].dt.hour
        df_bus_card["기준분"] = df_bus_card["승차시각"].dt.minute

        # "버스내체류시간(분)" 컬럼 추가
        df_bus_card["버스내체류시간(분)"] = \
            round((df_bus_card["하차시각"] - df_bus_card["승차시각"]).dt.total_seconds() / 60, 2)

        # 승하차구간 컬럼 추가하기
        df_bus_card["승하차구간"] = \
            df_bus_card_kor["승차정류장"] + " -> " + df_bus_card_kor["하차정류장"]

        # 원본 df 데이터프레임에 위 데이터프레임 행단위 추가하기
        df_bus_card_tot = pd.concat([df_bus_card_tot, df_bus_card], ignore_index=True)

    except : 
        print("오류가 발생했습니다.")
        break
```
