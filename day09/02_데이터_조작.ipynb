```python
### 데이터처리 라이브러리
import pandas as pd

### 시각화 라이브러리 
# - seaborn 라이브러는 샘플 데이터셋(행렬 데이터)을 제공함
import seaborn as sns
```

```python
### seaborn에서 제공해 주는 데이터셋 중 한개 불러들이기
# - 데이터셋 이름 : tips(팁 데이터셋)
# - 불러들이면 -> 데이터프레임 형태
tips = sns.load_dataset("tips")
tips.info()

### category 데이터타입 
#    - 기본적으로 문자열(object) 타입을 나타냄
#    - 범주형으로 분류 가능한 데이터인 경우에 category 데이터 타입으로 인식하는 경우가 있음
#      -> 범주형 데이터 : 하나의 열 데이터에 -> 남자/여자와 같이 분류가 가능한 데이터
#    - 일반적으로 문자열(object) 타입으로 인식

### 범주형 데이터 : 분류를 통해 집계 가능한 데이터들
### 연속형 데이터 : 집계가 불가능한 데이터들로 주로 숫자값을 사용하게 되며,
#                : 나이, 키, 몸무게 등과 같이 흐름에 따라 증/감이 연속해서 발생하는 데이터들..

### 컬럼 정의
# - total_bill : 식사 금액
# - tip : 팁 금액(손님들이 건낸 팁 금액)
# - sex : 팁을 건낸 사람의 성별(남/여)
# - smoker : 팁을 건낸 사람의 흡연여부
# - day : 팁을 건낸 사람이 식당을 찾은 요일
# - time : 팁을 건낸 사람이 아침/점심/저녁 중에 언제 식당을 들렀는지
# - size : 팁을 건낸 사람과 함께 식사한 전체 인원수
```

```python
"""
- 지금까지 사용한 데이터프레임 이름 뒤에 대괄호 안에 사용한 행번호는
  --> 눈에 보이는 인덱스 값 입니다. (메모리 내에서 관리하는 인덱스 번호가 아님)
- 인덱스 값을 이용하는 경우에는 -> 실제 눈에 보이는 값(번호)만 사용 가능

- 데이터프레임 내에 행/렬, 또는 특정 행렬 값에 접근하여 수정 등을 수행할 경우에는
  --> 넘파이(Numpy)에서 제공하는 loc과 iloc를 이용해야 함
  --> loc와 iloc는 실제 메모리에 접근해서 사용하는 방식
  --> 특정 컬럼의 값을 수정하고자 할때 주로 loc 또는 iloc로 접근하여 사용
  --> 이외 일반적으로 데이터 조회 시에도 사용
  
- loc[] : 인덱스 값을 이용하여 접근하는 방식(눈에 보이는 값)
- iloc[] : 실제 메모리에서 관리하는 인덱스 번호로 접근하는 방식(메모리 번호)
- Numpy 라이브러리 : 숫치 데이터 처리(연산 등 처리)를 위한 특화된 라이브러리
"""
```

```python
"""
- 새로운 컬럼 sex_no 추가
- sex_no에 사용할 데이터는 
  --> sex의 각 행의 값이 남성이면 1로, 여성이면 0으로 sex_no에 넣어주시면 됩니다.
"""
```

```python
### 열을 기준값 한개로 만들어 놓기
tips["sex_no"] = 0

### 남성의 값만 1로 수정
for idx in range(0, len(tips), 1) :
    if tips["sex"][idx] == "Male" :
        tips.iloc[idx, 8] = 1

tips
```

```python
### 열을 기준값 한개로 만들어 놓기
tips["sex_no"] = 0

### 남성의 값만 1로 수정
c = tips["sex"] == "Male"
tips.loc[c, "sex_no"] = 1

tips.loc[tips["sex"] == "Male", "sex_no"] = 1
tips
```

```python
### new_tip 컬럼 생성
# - new_tip에 사용할 데이터 값은 
#  -> total_bill의 값과 tip의 곱한 값을 각각 사용
tips["new_tip"] = tips["total_bill"] * tips["tip"]
tips
```

### 빈도 확인하기

```python
### 빈도는 범주형 데이터에 대한 간단한 집계를 할 때 사용
tips["sex"].value_counts()
```

```python
### 간단히 집계하는 방법
tips[["sex", "day"]]

### 성별 및 요일별로 집계하기
# - crosstab(인덱스 영역, 컬럼 영역)
#  --> 피봇 테이블의 결과와 유사
pd.crosstab(tips["sex"], tips["day"])
```

```python
### 성별 및 요일별로 집계하기
# - crosstab(인덱스 영역, 컬럼 영역)
#  --> 피봇 테이블의 결과와 유사
#  --> margins=True : 행단위, 열단위 전체 값들을 sum한 값이 All이란 이름으로 생성됨
pd.crosstab(tips["sex"], tips["day"], margins=True)
```
