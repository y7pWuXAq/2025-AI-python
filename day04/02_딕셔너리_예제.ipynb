#### 이름, 나이, 지역 데이터 관리하기
- 이름, 나이, 지역은 딕셔너리 타입 사용
- 리스트에 데이터 추가 시 {name : '이름', age : '나이', area : '지역}

- 리스트 변수명 : data_list
- [ {name : '이름', age : '나이', area : '지역} ... {name : '이름', age : '나이', area : '지역} ]


```python
### 임의의 값 입력 안내

msg = """
--- *** ---
메뉴 (번호로 입력)
    
0. 종료
1. 입력값 추가
2. 입력값 확인
3. 입력값 검색
4. 입력값 삭제
5. 입력값 수정
--- ** ---
"""

# 관리할 리스트
data_list = [ ]
```

```python

while True:
    print(msg)
    print(" ")
    menw = int(input("번호로 메뉴 선택 : "))

    # 종료 메뉴 선택
    if menw == 0 :
        print("입력이 종료 되었습니다.")
        break
    
    # 정보 입력 선택
    elif menw == 1 :
        print(" -- *** -- ")
        print('신규 등록을 위해 정보를 입력하세요.')
        print(' ')
        print("이름을 입력하세요")
        p_name = input("이름 : ")

        s_name = False # 이름 중복이 아닌 경우 

        # 이름 중복 확인
        for person in data_list :
            if person['name'] == p_name :
                print(f"'{p_name}'님은 이미 등록된 이름입니다. 다른 이름을 입력해주세요.")
                s_name = True # 중복일 때 True로

        if s_name == False :
            # 이름 중복이 없을 때
            # 이름이 중복되지 않으면 나이와 지역 입력
            print("나이를 입력하세요.")
            p_age = input("나이 : ")

            print("지역을 입력하세요.")
            print(' ')
            p_area = input("지역 : ")

            # 입력값을 딕셔너리로 만들어 리스트에 추가
            person = {'name' : p_name, 'age' : p_age, 'area' : p_area}
            data_list.append(person)

            print(f"{p_name}님의 정보가 추가되었습니다.")
            print(' ')
            print("안내 : 메뉴로 이동합니다.")
            print(" -- *** -- ")

    # 정보 확인 선택
    elif menw == 2 :
        print(" -- *** -- ")
        print("저장된 정보 목록입니다.")
        print(' ')

        if not data_list :
            print("등록된 정보가 없습니다.")
        else:
            for person in data_list :
                print(f"이름: {person['name']}, 나이: {person['age']}, 지역: {person['area']}")

        print(' ')
        print("안내 : 메뉴로 이동합니다.")
        print(" -- *** -- ")

    # 정보 검색 선택
    elif menw == 3 :
        print("검색 할 이름을 입력하세요.")
        s_name = input("이름 입력 : ")
        print(" ")

        # 리스트 내 이름 검색 후 해당 값 출력
        for person in data_list :
            if person['name'] == s_name :
                print(f"'{s_name}'님의 정보 확인.")
                print(f"이름: {person['name']}, 나이: {person['age']}, 지역: {person['area']}")
                print(' ')
                print("안내 : 메뉴로 이동합니다.")
                print(" -- *** -- ")
                break
            else :
                print("등록된 정보가 존재하지 않습니다.")
                print(' ')
                print("안내 : 메뉴로 이동합니다.")
                print(" -- *** -- ")
                break

    # 정보 삭제 선택
    elif menw == 4 :
        print("삭제할 이름을 입력하세요")
        p_name = input("이름 : ")

        s_name = True  # 검색한 이름이 정보에 있을 경우

        # 이름 중복 확인
        for person in data_list :
            if person['name'] == p_name :
                print(f"'{p_name}'님의 정보를 삭제하시겠습니까?")
                ans = input("삭제하시겠습니까? (y / n) : ")
                print(' ')

                if ans.lower() == 'y' :
                    data_list.remove(person)  # 이름과 일치하는 정보 삭제
                    s_name = False  # 삭제 후 False로
                    print(f"'{p_name}'님의 정보가 정상적으로 삭제되었습니다.")
                    print("안내 : 메뉴로 이동합니다.")
                    print(" -- *** -- ")
                    break  # 삭제 후 루프를 종료

        if s_name :  # 정보가 없을 경우
            print('등록된 정보가 존재하지 않습니다.')
            print(' ')
            print("안내 : 메뉴로 이동합니다.")
            print(" -- *** -- ")

    # 정보 수정 선택
    elif menw == 5 :
        print("검색할 이름을 입력하세요")
        p_name = input("이름 : ")

        s_name = True  # 이름 중복인 경우

        # 이름 중복 확인
        for person in data_list :
            if person['name'] == p_name :
                s_name = False  # 이름이 존재하는 경우

                print(f"'{p_name}'님의 정보를 수정합니다.")
                print(' ')
                print("어떤 항목을 수정하시겠습니까?")
                print("1. 이름")
                print("2. 나이")
                print("3. 지역")
                print("4. 전부 수정")
                print(' ')
                choice = input("수정할 항목 번호를 입력하세요 : ")
                print(' ')

                if choice == '1' :
                    new_name = input("새로운 이름을 입력하세요 : ")
                    person['name'] = new_name
                    print(f"이름이 '{new_name}'로 수정되었습니다.")
                    print(' ')

                elif choice == '2' :
                    new_age = input("새로운 나이를 입력하세요 : ")
                    person['age'] = new_age
                    print(f"나이가 '{new_age}'로 수정되었습니다.")
                    print(' ')

                elif choice == '3' :
                    new_region = input("새로운 지역을 입력하세요 : ")
                    person['region'] = new_region
                    print(f"지역이 '{new_region}'으로 수정되었습니다.")
                    print(' ')

                elif choice == '4' :
                    new_name = input("새로운 이름을 입력하세요 : ")
                    new_age = input("새로운 나이를 입력하세요 : ")
                    new_region = input("새로운 지역을 입력하세요 : ")
                    print(' ')

                    person['name'] = new_name
                    person['age'] = new_age
                    person['region'] = new_region
                    print(f"수정되었습니다 : 이름 : {new_name}, 나이 : {new_age}, 지역 : {new_region}")
                    print(' ')

                else:
                    print("잘못된 번호를 입력하셨습니다.")
                    print(' ')

                print("수정이 완료되었습니다.")
                print("안내 : 메뉴로 이동합니다.")
                print(" -- *** -- ")
                break  # 수정 후 루프 종료

        if s_name :  # 이름이 존재하지 않으면
            print('등록된 정보가 존재하지 않습니다.')
            print(' ')
            print("안내 : 메뉴로 이동합니다.")
            print(" -- *** -- ")

    else :
        print("입력 번호를 확인해주세요.")
        print("메뉴로 이동합니다.")
        print(" -- *** -- ")
        continue
```
