```python
import pandas as pd

# 통합할 데이터프레임 정의
df = pd.DataFrame()

# 파일 경로 리스트 작성
read_file_list = []  # 파일 경로를 담을 리스트

for y in range(2010, 2020):
    for m in range(1, 13):
        if y == 2019 and m == 9:
            break
        yearmonth = f"{y}{m:02d}"
        file_path_total = f"./files2/kto_{yearmonth}.xlsx"
        read_file_list.append(file_path_total)  # 파일 경로를 리스트에 추가

# 대륙 리스트 정의
continents = ["아시아"]*25 + ["아메리카"]*5 + ["유럽"]*23 + ["오세아니아"]*3 + \
                ["아프리카"]*2 + ["기타대륙"] + ["교포"]

# 파일 읽고 처리한 후, df에 누적 추가하는 과정
for file_path in read_file_list:
    # 파일 읽기
    kto_data = pd.read_excel(file_path,
                             header=1,
                             skipfooter=4,
                             usecols="A:G")

    # 기준년월 컬럼 추가하기
    yearmonth = file_path.split('_')[-1].split('.')[0]  # 파일명에서 연월 추출
    kto_data["기준년월"] = f"{yearmonth[:4]}-{yearmonth[4:]}"
    
    # 실제 국적만 추출하기
    c = kto_data["국적"].isin(['아시아주', '미주', '구주', '대양주', '아프리카주', '기타대륙', '교포소계'])
    kto_data_country = kto_data[c == False]
    
    # 인덱스 재정의
    kto_data_country_newindex = kto_data_country.reset_index(drop=True)
    
    # 대륙 컬럼 추가
    kto_data_country_newindex["대륙"] = continents
    
    # 관광객비율(%) 컬럼 추가
    kto_data_country_newindex["관광객비율(%)"] = round(kto_data_country_newindex["관광"] / kto_data_country_newindex["계"] * 100, 3)
    
    # 전체 관광객비율(%) 계산
    total_tourism_sum = kto_data_country_newindex["관광"].sum()
    kto_data_country_newindex["전체관광객비율(%)"] = round(kto_data_country_newindex["관광"] / total_tourism_sum * 100, 3)

    # 각 파일의 데이터를 df에 누적 추가
    df = pd.concat([df, kto_data_country_newindex], ignore_index=True)

# 통합된 데이터프레임을 엑셀 파일로 저장
df.to_excel("rto_total.xlsx", index=False)
```
