```python
### 예제

# - 리스트 변수 생성 : 이름, 나이, 주소 각각의 새 리스트 생성
# - 리스트 변수명 : name, age, area

# - 데이터 추가 : "홍길동", 10, "부산"
# - 데이터 추가 : "이순신", 20, "서울"

# 처리 방법 
# - 무한 반복
# - 임의 번호를 입력 받기
# - 입력 받은 번호가 1이면 데이터 추가를 수행
# - 데이터 추가 시 이름, 나이, 지역을 각각 입력 받기
# - 입력 받은 각각의 이름, 나이, 지역의 값을 리스트에 추가
# - 한번 입력이 끝나면 다시 임의 번호를 입력
# - 추가 할 데이터가 없을 경우 임의 번호로 무한 반복 종료

# 안내 메세지
# - 임의 번호를 입력 받기 
# - 이외 데이터 추가시 입력 값에 대한 안내는 직접 작성
```

```python
## 리스트 변수
name = []
age = []
area = []

### 임의의 값 입력 안내
msg = """
--- *** ---
메뉴 선택 (번호로 입력)

0. 종료
1. 입력값 추가
2. 입력값 확인
3. 입력값 삭제
4. 입력값 수정
--- ** ---
"""

# print(name)
```

```python
while True :
    print(msg)
    menw = int(input("메뉴를 선택하세요 : "))

    if menw == 0 :
        print("입력이 종료 되었습니다.")
        break

    elif menw == 1 :
        print("이름을 입력하세요")
        name.append(input("이름 : "))

        print("나이를 입력하세요")
        age.append(input("나이 : "))

        print("지역을 입력하세요")
        area.append(input("지역 : "))

        print("정상적으로 추가 되었습니다.")

    elif menw == 2 :
        print("저장된 값입니다.")
        print(name)
        print(age)
        print(area)

    else :
        print("입력 번호를 확인해주세요.")
        continue

```

```python
### 특정 값이 리스트 내에 포함되어 있는지 확인하기

list2 = ["홍길동","이순신", "김유신"]

name1 = "이순신"
# 이순신이라는 값이 리스트 내에 포함되어 있는지 확인
# 리스트 앞쪽에 in을 붙히면 왼쪽의 값이 오른쪽 리스트 내에 포함되어 있는지 확인
# 포함 되어 있으면 True, 없으면 False

# in : 포함 관계 여부를 나타내는 조건 연산

if name1 in list2 :
    print(f"{name1}이 포함 되어 있습니다.")
else :
    print(f"{name1}이 포함 되어 있지 않습니다.")

# 포함이 안된 경우를 조건 처리 할 경우에는 not in
if name1 not in list2 :
    print(f"{name1}이 포함 되어 있지 않습니다.")
else :
    print(f"{name1}이 포함 되어 있습니다.")
```
