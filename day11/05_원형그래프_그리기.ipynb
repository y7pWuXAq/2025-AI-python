```python
### 데이터 처리
import pandas as pd

### 시각화 처리
# - 파이썬에서 사용되는 기본 시각화 라이브러리
import matplotlib.pyplot as plt

# - 히트맵 라이브러리
import seaborn as sns

# - 한글처리
plt.rc("font", family="NanumGothic")

# 마이너스 기호 깨짐 처리
plt.rcParams["axes.unicode_minus"] = False 
```

```python
### 사용할 데이터
# - label 은 데이터 프레임의 특정 컬럼명이라고 생각
label = ["A", "B", "C", "D"]

# 해당 특정 컬럼의 범주별 집계 데이터라고 생각
sizes = [25, 30, 20, 25]

## 원형 차트 그리기
# - pie()

### 대표 그래프 관리자 정의
# - 별도의 작은 그래프 공간이 없는 상태로 설정 (Axes는 1개의 공간만 사용 됨)
fig, ax = plt.subplots()

### 하나의 공간에 원형 차트 그리기
# - 첫번째 : 빈도 데이터
# - 두번째 : 표시할 범주 이름
# - autopct : 빈도 데이터를 표시(기호% 사용), 1.1f는 소숫점 자릿수 지정

# - wedges : 원형 차트 자체 정보 반환
# - texts : label 정보 반환(label 표시 스타일 저징시 사용)
# 
wedges, texts, autotexts = ax.pie(sizes, labels=label, autopct="%1.1f%%", startangle=45)

# autotexts (집계(빈도) 데이터) 텍스트 스타일 지정
plt.setp(autotexts, size = 8, weight = "bold", color="white", va="center")

# texts(label)에 대한 텍스트 스타일 지정
# - 텍스트 스타일 지정
plt.setp(texts, size=10, weight="bold", color="black", va="center")

plt.show()
```
