```python
### 데이터 처리
import pandas as pd

### 시각화 처리
# - 파이썬에서 사용되는 기본 시각화 라이브러리
import matplotlib.pyplot as plt

# - 히트맵 라이브러리
import seaborn as sns

# - 한글처리
plt.rc("font", family="NanumGothic")

# 마이너스 기호 깨짐 처리
plt.rcParams["axes.unicode_minus"] = False 
```

```python
### 샘플데이터

x = [1, 2, 3, 4, 5]
y = [100, 200, 300, 400, 500]

### bar(막대) 그래프 그리기
plt.bar(x,y,
        # - edgecolor : 막대 그래프 테두리 색상 지정
        edgecolor="black",
        # - 투명도 : 막대 그래프 내부 투명도 0~1 사이 값 사용
        alpha=0.5,
        # - 막대 그래프 각각의 색 지정
        color=["skyblue", "yellow", "green", "red", "purple"]
        )

### 각 막대별로 x축 및 y축의 위치에 value 값 표시하기
# - enumerate() : 리스트의 인덱스와 각 인덱스의 값을 튜플로 세트를 만들어주는 함수
# [ (인덱스 번호, value의 값), ( ), ( ) , ... ]
# [ (0, 1), (1, 2), (2, 3), (3, 4), (4, 5) ]
# - zip() 함수와 동일한 결과 형태로 반환

for idx, x_v in enumerate(x) :
    # 각 막대에 y축의 값 표시
    plt.text(
        # 텍스트가 위치할 x축 위치 값 정의 (x 좌표)
        x_v,
        # 텍스트가 위치할 y축 위치 값 정의 (y 좌표)
        y[idx],
        # 텍스트로 표시할 값 정의
        y[idx],
        # 텍스트 color(색상) 정의
        color="black",
        # 텍스트 정렬
        # - left, center, right
        horizontalalignment="center", # 수평
        # - top, center, bottom
        verticalalignment = "bottom" # 수직
    )

plt.show()
```

```python
# 위 예제를 그대로 사용
# 각 막대 텍스트의 색상을 막대의 색상과 동일하게 처리

### 샘플데이터

x = [1, 2, 3, 4, 5]
y = [100, 200, 300, 400, 500]

# 막대 색상 배열
bar_colors = ["skyblue", "yellow", "green", "red", "purple"]

### bar(막대) 그래프 그리기
plt.bar(x,y,
        # - edgecolor : 막대 그래프 테두리 색상 지정
        edgecolor="black",
        # - 투명도 : 막대 그래프 내부 투명도 0~1 사이 값 사용
        alpha=0.5,
        # - 막대 그래프 각각의 색 지정
        color=bar_colors
        )

for idx, x_v in enumerate(x) :
    # 각 막대에 y축의 값 표시
    plt.text(
        # 텍스트가 위치할 x축 위치 값 정의 (x 좌표)
        x_v,
        # 텍스트가 위치할 y축 위치 값 정의 (y 좌표)
        y[idx],
        # 텍스트로 표시할 값 정의
        y[idx],
        # 텍스트 color(색상) 정의: 막대 색상과 동일하게 설정
        color=bar_colors[idx],  # 각 막대의 색상 가져오기
        # 텍스트 정렬
        # - left, center, right
        horizontalalignment="center", # 수평
        # - top, center, bottom
        verticalalignment = "bottom" # 수직
    )

plt.show()
```

```python
# 예제 이어서
# 값(y)이 가장 큰 텍스트의 색상은 빨강색으로 나머지는 검은색으로 처리

### 샘플데이터

x = [1, 2, 3, 4, 5]
y = [100, 200, 300, 400, 500]

# 막대 색상 배열
bar_colors = ["skyblue", "yellow", "green", "red", "purple"]

# y값 중 가장 큰 값의 인덱스 찾기
max_y_index = y.index(max(y))

### bar(막대) 그래프 그리기
plt.bar(x,y,
        # - edgecolor : 막대 그래프 테두리 색상 지정
        edgecolor="black",
        # - 투명도 : 막대 그래프 내부 투명도 0~1 사이 값 사용
        alpha=0.5,
        # - 막대 그래프 각각의 색 지정
        color=bar_colors
        )

for idx, x_v in enumerate(x):
    # 텍스트 color를 조건에 따라 설정
    if idx == max_y_index : text_color = "red" 
    else : text_color = "black"

    # 각 막대에 y축의 값 표시
    plt.text(
        # 텍스트가 위치할 x축 위치 값 정의 (x 좌표)
        x_v,
        # 텍스트가 위치할 y축 위치 값 정의 (y 좌표)
        y[idx],
        # 텍스트로 표시할 값 정의
        y[idx],
        # 텍스트 color(색상) 정의: 막대 색상과 동일하게 설정
        color = text_color,
        # 텍스트 정렬
        # - left, center, right
        horizontalalignment="center", # 수평
        # - top, center, bottom
        verticalalignment = "bottom" # 수직
    )

plt.show()
```

```python
### 그래프 내 화살표 넣기

# 예제 이어서
# 값(y)이 가장 큰 텍스트의 색상은 빨강색으로 나머지는 검은색으로 처리

### 샘플데이터

x = [1, 2, 3, 4, 5]
y = [100, 200, 300, 400, 500]

# 막대 색상 배열
bar_colors = ["skyblue", "yellow", "green", "red", "purple"]

# y값 중 가장 큰 값의 인덱스 찾기
max_y_index = y.index(max(y))

### bar(막대) 그래프 그리기
plt.bar(x,y,
        # - edgecolor : 막대 그래프 테두리 색상 지정
        edgecolor="black",
        # - 투명도 : 막대 그래프 내부 투명도 0~1 사이 값 사용
        alpha=0.5,
        # - 막대 그래프 각각의 색 지정
        color=bar_colors
        )

for idx, x_v in enumerate(x):
    # 텍스트 color를 조건에 따라 설정
    if idx == max_y_index : text_color = "red" 
    else : text_color = "black"

    # 각 막대에 y축의 값 표시
    plt.text(
        # 텍스트가 위치할 x축 위치 값 정의 (x 좌표)
        x_v,
        # 텍스트가 위치할 y축 위치 값 정의 (y 좌표)
        y[idx],
        # 텍스트로 표시할 값 정의
        y[idx],
        # 텍스트 color(색상) 정의: 막대 색상과 동일하게 설정
        color = text_color,
        # 텍스트 정렬
        # - left, center, right
        horizontalalignment="center", # 수평
        # - top, center, bottom
        verticalalignment = "bottom" # 수직
    )

### 그래프 내 화살표 표시
# - 함수 : plt.annotate(표시할 텍스트, 화살표 시작위치, 끝위치, 그리기 옵션, 특성설정) 
# - 표시할 텍스트 : 화살표 선 위에 텍스트를 넣고자 할 때 사용, 텍스트 없을 경우 공백("")
plt.annotate(
    # 표시할 텍스트 없으면 공백
    "",
    # 화살표 시작 위치 (x, y) 좌표
    xytext=(0.8, 230),
    # 화살표 끝나는 위치 (x, y) 좌표
    xy=(3.8, 500),
    # 화살표 그리기 옵션
    xycoords='data',
    # 화살표 특성 설정
    # - arrowstyle : 화살표 스타일
    # - colro : 색상
    # - lw :  선 굵기
    arrowprops=dict(arrowstyle='->', color='olive', lw=3)
    )


plt.show()
```
